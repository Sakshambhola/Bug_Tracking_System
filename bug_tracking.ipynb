```python

class Employee:
    def __init__(self, employee_id, name):
        self.employee_id = employee_id
        self.name = name

class Customer:
    def __init__(self, customer_id, name):
        self.customer_id = customer_id
        self.name = name

class Bug:
    def __init__(self, bug_id, description, status):
        self.bug_id = bug_id
        self.description = description
        self.status = status





def login():
    # Prompt the user for credentials
    username = input("Enter your username: ")
    password = input("Enter your password: ")

    # Validate the credentials
    if validate_credentials(username, password):
        print("Login successful!")
        # Perform further actions or redirect to the main system
    else:
        print("Invalid username or password. Please try again.")

def validate_credentials(username, password):
    # Check if the username and password match the stored credentials
    # This can be replaced with your actual validation logic (e.g., querying a database)
    stored_username = "admin"
    stored_password = "password"

    if username == stored_username and password == stored_password:
        return True
    else:
        return False

# Call the login function to start the authentication process
login()


class Customer:
    def __init__(self, name, login_id):
        self.name = name
        self.login_id = login_id

class AdminModule:
    def __init__(self):
        self.customers = []

    def add_customer(self, name, login_id):
        customer = Customer(name, login_id)
        self.customers.append(customer)
        print("Customer added successfully.")

    def view_all_customers(self):
        print("All Customers:")
        for customer in self.customers:
            print("Name: {}, Login ID: {}".format(customer.name, customer.login_id))

    def search_by_name(self, name):
        print("Search Results by Name:")
        for customer in self.customers:
            if customer.name.lower() == name.lower():
                print("Name: {}, Login ID: {}".format(customer.name, customer.login_id))

    def search_by_login_id(self, login_id):
        print("Search Results by Login ID:")
        for customer in self.customers:
            if customer.login_id.lower() == login_id.lower():
                print("Name: {}, Login ID: {}".format(customer.name, customer.login_id))

# Create an instance of the AdminModule
admin = AdminModule()

# Add some customers for testing
admin.add_customer("John Doe", "john123")
admin.add_customer("Jane Smith", "jane456")

# View all customers
admin.view_all_customers()

# Search by customer name
admin.search_by_name("John Doe")

# Search by customer login ID
admin.search_by_login_id("jane456")


class Employee:
    def __init__(self, name, employee_id, employee_type, is_active=True):
        self.name = name
        self.employee_id = employee_id
        self.employee_type = employee_type
        self.is_active = is_active

class EmployeeServices:
    def __init__(self):
        self.employees = []

    def add_employee(self, name, employee_id, employee_type):
        employee = Employee(name, employee_id, employee_type)
        self.employees.append(employee)
        print("Employee added successfully.")

    def view_all_employees(self):
        print("All Employees:")
        for employee in self.employees:
            print("Name: {}, Employee ID: {}, Type: {}, Active: {}".format(
                employee.name, employee.employee_id, employee.employee_type, employee.is_active
            ))

    def search_by_name(self, name):
        print("Search Results by Name:")
        for employee in self.employees:
            if employee.name.lower() == name.lower():
                print("Name: {}, Employee ID: {}, Type: {}, Active: {}".format(
                    employee.name, employee.employee_id, employee.employee_type, employee.is_active
                ))

    def search_by_employee_id(self, employee_id):
        print("Search Results by Employee ID:")
        for employee in self.employees:
            if employee.employee_id.lower() == employee_id.lower():
                print("Name: {}, Employee ID: {}, Type: {}, Active: {}".format(
                    employee.name, employee.employee_id, employee.employee_type, employee.is_active
                ))

    def edit_employee(self, employee_id, new_name):
        for employee in self.employees:
            if employee.employee_id.lower() == employee_id.lower():
                employee.name = new_name
                print("Employee details updated successfully.")
                return
        print("Employee not found.")

    def activate_deactivate_employee(self, employee_id, is_active):
        for employee in self.employees:
            if employee.employee_id.lower() == employee_id.lower():
                employee.is_active = is_active
                print("Employee status updated successfully.")
                return
        print("Employee not found.")

# Create an instance of the EmployeeServices
employee_services = EmployeeServices()

# Add some employees for testing
employee_services.add_employee("John Doe", "EMP001", "Admin")
employee_services.add_employee("Jane Smith", "EMP002", "Manager")
employee_services.add_employee("David Johnson", "EMP003", "Expert")

# View all employees
employee_services.view_all_employees()

# Search by employee name
employee_services.search_by_name("John Doe")

# Search by employee ID
employee_services.search_by_employee_id("EMP002")

# Edit employee details
employee_services.edit_employee("EMP003", "David Anderson")

# Activate/deactivate employee
employee_services.activate_deactivate_employee("EMP001", False)
employee_services.view_all_employees()



class Customer:
    def __init__(self, name, login_id, password):
        self.name = name
        self.login_id = login_id
        self.password = password
        self.bugs = []

    def update_account(self, name):
        self.name = name
        print("Account updated successfully.")

    def post_bug(self, bug):
        self.bugs.append(bug)
        print("Bug posted successfully.")

    def view_all_bugs(self):
        print("All Bugs:")
        for bug in self.bugs:
            print("Bug ID: {}, Status: {}, Description: {}".format(
                bug.bug_id, bug.status, bug.description
            ))

    def search_bug_by_status(self, status):
        print("Search Results by Status ({}):".format(status))
        for bug in self.bugs:
            if bug.status.lower() == status.lower():
                print("Bug ID: {}, Status: {}, Description: {}".format(
                    bug.bug_id, bug.status, bug.description
                ))

    def view_bug_by_status(self, status):
        print("Viewing Bugs with Status ({}):".format(status))
        for bug in self.bugs:
            if bug.status.lower() == status.lower():
                print("Bug ID: {}, Status: {}, Description: {}".format(
                    bug.bug_id, bug.status, bug.description
                ))
                if bug.solution:
                    print("Solution: {}".format(bug.solution))

    def change_password(self, current_password, new_password):
        if self.password == current_password:
            self.password = new_password
            print("Password changed successfully.")
        else:
            print("Invalid current password. Please try again.")

class Bug:
    def __init__(self, bug_id, status, description, solution=None):
        self.bug_id = bug_id
        self.status = status
        self.description = description
        self.solution = solution

# Create a customer instance
customer = Customer("John Doe", "john123", "password123")

# Update account
customer.update_account("John Smith")

# Post new bug
bug1 = Bug(1, "Open", "Application crashes on startup")
customer.post_bug(bug1)

# View all bugs
customer.view_all_bugs()

# Search bugs by status
customer.search_bug_by_status("Open")

# View bugs by status
customer.view_bug_by_status("Open")

# Change password
customer.change_password("password123", "newpassword123")



class Expert:
    def __init__(self, name, login_id, password):
        self.name = name
        self.login_id = login_id
        self.password = password
        self.assigned_bugs = []

    def view_assigned_bugs(self):
        print("Assigned Bugs:")
        for bug in self.assigned_bugs:
            print("Bug ID: {}, Status: {}, Description: {}".format(
                bug.bug_id, bug.status, bug.description
            ))

    def filter_assigned_bugs_by_status(self, status):
        print("Filtering Assigned Bugs by Status ({}):".format(status))
        filtered_bugs = [bug for bug in self.assigned_bugs if bug.status.lower() == status.lower()]
        for bug in filtered_bugs:
            print("Bug ID: {}, Status: {}, Description: {}".format(
                bug.bug_id, bug.status, bug.description
            ))

    def solve_bug(self, bug_id, solution):
        for bug in self.assigned_bugs:
            if bug.bug_id == bug_id:
                bug.status = "Solved"
                bug.solution = solution
                print("Bug solved successfully.")
                return
        print("Bug not found.")

    def change_password(self, current_password, new_password):
        if self.password == current_password:
            self.password = new_password
            print("Password changed successfully.")
        else:
            print("Invalid current password. Please try again.")

class Bug:
    def __init__(self, bug_id, status, description, solution=None):
        self.bug_id = bug_id
        self.status = status
        self.description = description
        self.solution = solution

# Create an expert instance
expert = Expert("Jane Smith", "jane123", "password123")

# View assigned bugs
expert.view_assigned_bugs()

# Filter assigned bugs by status
expert.filter_assigned_bugs_by_status("Open")

# Solve a bug
bug1 = Bug(1, "Open", "Application crashes on startup")
expert.assigned_bugs.append(bug1)  # Assign the bug to the expert
expert.solve_bug(1, "Fixed a memory leak")

# Change password
expert.change_password("password123", "newpassword123")
```
